## Plot tracking results

```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
# load data:
with open('track.json', 'r') as f:
     track = json.load(f)
```

```python
track.keys()
```

```python
np.shape(track['zeta'])
```

```python
zeta  = np.array(track['zeta'])
x  = np.array(track['x'])
y  = np.array(track['y'])
delta = np.array(track['delta'])
p0c   = np.array(track['p0c'])
pID   = np.array(track['pID'])
#parent_pID = np.array(track['parent_pID'])
state = np.array(track['state'])
```

```python
plt.figure(dpi=100)
plt.grid()

N_turns = len(zeta[0])

turn = 0
plt.scatter(100*zeta[:,turn],100*delta[:,turn], label='turn %g' % turn, s=2)

turn = int(N_turns/2)
plt.scatter(100*zeta[:,turn],100*delta[:,turn], label='turn %g' % turn, s=2)

turn = N_turns-1
plt.scatter(100*zeta[:,turn],100*delta[:,turn], label='turn %g' % turn, s=2)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.legend()
plt.ylim(-0.1,+0.1)

#plt.savefig('image.png')
plt.show()
```

## Excited ions:

```python
plt.figure(dpi=100)
plt.grid()

N_turns = len(zeta[0])

turn = int(N_turns/2)
z = zeta[:,turn]*100; dp = delta[:,turn]*100; excited = (state[:,turn] == 2)
plt.scatter(z,dp, label='Still', s=2, alpha=0.3)
plt.scatter(z[excited],dp[excited], label='Excited', s=3, c='black', alpha=0.6)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.legend()
plt.ylim(-0.1,+0.1)

#plt.savefig('image.png')
plt.show()
```

```python
plt.figure(dpi=100)
plt.grid()

z = zeta[:,turn]*100; y_mm = y[:,turn]*1e3; excited = (state[:,turn] == 2)
plt.scatter(z,y_mm, label='Still', s=2, alpha=0.3)
plt.scatter(z[excited],y_mm[excited], label='Excited', s=3, c='black', alpha=0.6)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$y$ (mm)')

plt.legend()
#plt.ylim(-0.1,+0.1)

#plt.savefig('image.png')
plt.show()
```
