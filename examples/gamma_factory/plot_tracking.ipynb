## Plot tracking results

```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
# load data:
with open('track.json', 'r') as f:
     track = json.load(f)
```

```python
track.keys()
```

```python
np.shape(track['zeta'])
```

```python
turns  = np.array(track['turns'])
zeta  = np.array(track['zeta'])
x  = np.array(track['x'])
y  = np.array(track['y'])
delta = np.array(track['delta'])
p0c   = np.array(track['p0c'])
pID   = np.array(track['pID'])
#parent_pID = np.array(track['parent_pID'])
state = np.array(track['state'])
```

```python
plt.figure(dpi=150)

plt.subplot(2,1,1)

N_saves = len(zeta)

save = 0
plt.scatter(100*zeta[save],100*delta[save],
            label='turn %g' % turns[save], s=1, alpha=0.1)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.grid()
plt.xlim(-30,+30)
plt.ylim(-0.3,+0.3)
plt.title(f'Turn {turns[save]}')

#save = int(N_saves/2)
#plt.scatter(100*zeta[save],100*delta[save], label='turn %g' % turns[save], s=1)

plt.subplot(2,1,2)

save = -1
plt.scatter(100*zeta[save],100*delta[save],
            label='turn %g' % turns[save], s=1, alpha=0.1)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.grid()
plt.xlim(-30,+30)
plt.ylim(-0.3,+0.3)
plt.title(f'Turn {turns[save]}')

#plt.savefig('image.png')
plt.tight_layout()
plt.show()
```

## Excited ions:

```python
plt.figure(dpi=100)
plt.grid()

save = -1

z = zeta[save]*100; dp = delta[save]*100; excited = (state[save] == 2)
plt.scatter(z,dp, label='Still', s=2, alpha=0.1)
plt.scatter(z[excited],dp[excited], label='Excited', s=3, c='black', alpha=0.3)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.legend()
plt.ylim(-0.1,+0.1)

title = "%.1f %% of ions excited" % (100.0*len(z[excited])/len(z))
plt.title(title)

#plt.savefig('image.png')
plt.show()
```
