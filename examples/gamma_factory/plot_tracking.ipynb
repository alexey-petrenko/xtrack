## Plot tracking results

```python
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
# load data:
with open('track.json', 'r') as f:
     track = json.load(f)
```

```python
track.keys()
```

```python
np.shape(track['zeta'])
```

```python
zeta  = np.array(track['zeta'])
delta = np.array(track['delta'])
p0c   = np.array(track['p0c'])
pID   = np.array(track['pID'])
parent_pID = np.array(track['parent_pID'])
state = np.array(track['state'])
```

```python
plt.figure(dpi=100)
plt.grid()

N_turns = len(zeta[0])

turn = 0
plt.scatter(100*zeta[:,turn],100*delta[:,turn], label='turn %g' % turn)

turn = int(N_turns/2)
plt.scatter(100*zeta[:,turn],100*delta[:,turn], label='turn %g' % turn)

turn = N_turns-1
plt.scatter(100*zeta[:,turn],100*delta[:,turn], label='turn %g' % turn)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.legend()
plt.ylim(-0.012,+0.012)

#plt.savefig('image.png')
plt.show()
```

```python
plt.figure(dpi=100)
plt.grid()

N_turns = len(zeta[0])
turns = np.arange(N_turns)
pid = 8
plt.scatter(100*zeta[pid], 100*delta[pid], label='particle %g' % pid, s=2)

pid = 7
plt.scatter(100*zeta[pid], 100*delta[pid], label='particle %g' % pid, s=2)

plt.xlabel(r'$\zeta$ (cm)')
plt.ylabel(r'$\Delta p/p$ (%)')

plt.ylim(-0.012,+0.012)
plt.legend()

#plt.savefig('image.png')
plt.show()
```
